```python
def outlier_remover(data, columns):
    for col in columns:
        q1  = data[f'{col}'].quantile(0.25)
        q3  = data[f'{col}'].quantile(0.75)
        iqr = q3 - q1
        lower = q1 - 1.5*iqr
        upper = q3 + 1.5*iqr
        df1 = data[(data[f'{col}'] > lower) & (data[f'{col}'] < upper) ]
        return df1
```

```python
outlier_columns = ['Price', 'Levy', 'Mileage']
```

```python
df_new = outlier_remover(df, outlier_columns)
df_new.shape
```

### 5. Building Model

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X = df_new.drop('Price', axis = 1)
X.head()
```

```python
y = df_new.Price
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25)
```

#### One hot encoding

```python
ohe = OneHotEncoder()
ohe.fit(X[['Manufacturer', 'Model', 'Category','Leather interior', 'Fuel type', 'Gear box type']])
```

#### Column transformer for categorical values

```python
col_transformer=make_column_transformer((OneHotEncoder(categories=ohe.categories_), ['Manufacturer', 'Model', 'Category', 
                                    'Leather interior', 'Fuel type', 'Gear box type']), remainder='passthrough')
```

```python
lr  = LinearRegression()
dt  = DecisionTreeRegressor()
rf  = RandomForestRegressor()
xgb = XGBRegressor(max_depth = 7, n_estimators=500, learning_rate =.05)
cb  = CatBoostRegressor(learning_rate =.07, max_depth = 12, verbose=0)

regressors = [('Linear Regression', lr), ('Decision Tree', dt), ('Random Forest', rf), 
              ('XGBRegressor', xgb),('CatBoostRegressor', cb)]
```

#### Finding best model

```python
for regressor_name, regressor in regressors:
    
    pipe=make_pipeline(col_transformer,regressor)
    pipe.fit(X_train,y_train) 
    y_pred=pipe.predict(X_test)
    accuracy = np.round(r2_score(y_test,y_pred),2)
    
    print(f'{regressor_name} : {accuracy*100}%')
```

```python
pipe=make_pipeline(col_transformer,rf)
pipe.fit(X_train,y_train)
```
